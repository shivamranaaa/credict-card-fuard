```python
list_1=[i for i in df.columns]
list_1.remove("Class")
list_1.remove("Time")
list_1
```

```python
for i in list_1:
    iqr=df[f"{i}"].quantile(0.75)-df[f"{i}"].quantile(0.25)

    lower_boundary=df[f"{i}"].quantile(0.25)-(1.5*iqr)
    upper_boundary=df[f"{i}"].quantile(0.75)+(1.5*iqr)
    if lower_boundary>0:
        df.loc[df[f"{i}"]<=lower_boundary,f"{i}"]=lower_boundary
    elif upper_boundary>0:
        df.loc[df[f"{i}"]>=upper_boundary,f"{i}"]=upper_boundary
    else:
        pass
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
y.shape
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
lo= LogisticRegression().fit(X_train,y_train)
```

```python
y_pred=lo.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
